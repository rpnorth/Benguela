```python
# ====== load drifter derived Rossby number and divergence
DrifterClusters_list = []
drifter_release = np.array((1,2,3))

for dr in drifter_release:
    filenames = ("/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/Ro_Div_from_Release"+str(dr))
    print(filenames)
    matTri = scipy.io.loadmat(filenames+"triangles.mat") # check what variables are there (in the dictionary): 
    matLS = scipy.io.loadmat(filenames+"leastSquares.mat") # check what variables are there (in the dictionary): 
    #matLS.keys()

    Ro = np.array((matTri['Ro'],matLS['Ro'])).transpose(0,2,1)
    Div = np.array((matTri['Div'],matLS['Div'])).transpose(0,2,1)
    Lat = matTri['Lat'].transpose(1,0)
    Lon = matTri['Lon'].transpose(1,0)

    d_clust_array = np.arange(1,Ro.shape[1]+1,1) # number of clusters
    # get matlab days into datetime format
    date = [dt.datetime.fromordinal(int(ij)) + dt.timedelta(days=ij%1) - dt.timedelta(days = 366) 
            for ij in matTri['time'].flatten()] # Convert time axis to date form
    DrifterClusters_Ro_Div_Release = xr.Dataset({'RossbyNumber': (['calc_method','drifter_cluster','time'], 
                                                                   Ro),
                                                  'Divergence': (['calc_method','drifter_cluster','time'], 
                                                                 Div),
                                                  'lat': (['drifter_cluster','time'],Lat),
                                                  'lon': (['drifter_cluster','time'],Lon)},
                                                 coords={'time': np.asarray(date),
                                                         'drifter_cluster':d_clust_array,
                                                         'calc_method':list(('Triangle','LeastSquared'))})
    
    if dr==1:  DrifterClusters_Ro_Div_Release1 = DrifterClusters_Ro_Div_Release
    DrifterClusters_list.append(DrifterClusters_Ro_Div_Release)
    
DrifterClusters = xr.concat(DrifterClusters_list,dim='drifter_release').assign_coords(drifter_release=drifter_release)
DrifterClusters
```

```python
# ====== load drifter tracks, temperatures and velocities
# Load in file and call it mat
filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Data/MET_132/Drifters/Drifter_tracjectories.mat"
mat = scipy.io.loadmat(filenames) # check what variables are there (in the dictionary): 
mat.keys()
# get matlab days into datetime format
date_start = [dt.datetime.fromordinal(int(ij)) + dt.timedelta(days=ij%1) - dt.timedelta(days = 366) 
          for ij in mat['tstart'][0,:]] # Convert time axis to date form:

# a nested dictionary of xarray datasets, by drifter number; eventually add layer above for deployment.
drifters = OrderedDict() # Preallocate output dictionary
# Iterate through drifters in deployment 
# !! Removing last drifter as data seems to be missing
for drifter in range(mat['x'].shape[1]-1):
    # need to remove nan at the start to get time right
    mask = np.isnan(mat['u'][:,drifter])
    # create time vector based on deployment time, and dt = 30 min
    date_vec = np.arange(date_start[drifter],date_start[drifter]+dt.timedelta(hours=0.5)*mat['x'][~mask,drifter].shape[0],dt.timedelta(hours=0.5))
    drifters['Drifter{}'.format(drifter+1)] = xr.Dataset({'u': (['time'], mat['u'][~mask,drifter]),'v': (['time'], mat['v'][~mask,drifter]),
                                                         'lon': (['time'],mat['x'][~mask,drifter]),'lat':(['time'],mat['y'][~mask,drifter]),
                                                         'temperature': (['time'], mat['t'][~mask,drifter])},
                                       coords={'time': date_vec})
    drifters['Drifter{}'.format(drifter+1)].coords['z'] = 15
    #drifters['Drifter{}'.format(drifter+1)].coords['x_m'] = drifters['Drifter{}'.format(drifter+1)].coords['time']
    drifters['Drifter{}'.format(drifter+1)] = drifters['Drifter{}'.format(drifter+1)].assign_coords(x_m = ('time',np.append(0,np.cumsum(
                                                                                            gsw.distance(drifters['Drifter{}'.format(drifter+1)].lon.values,
                                                                                                         drifters['Drifter{}'.format(drifter+1)].lat.values,p=0)))))
    

# === Now add calculated variables from Jlab in matlab
# Load in file and call it mat
filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/ellipse_time_radius_velocity_Ro.mat"
mat = scipy.io.loadmat(filenames) # check what variables are there (in the dictionary): 
#filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/ellipse_lonres_latres.mat"
#mat2 = scipy.io.loadmat(filenames) # check what variables are there (in the dictionary): 
#filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/ellipse_curve_complex_for_plotting.mat"
#mat3 = scipy.io.loadmat(filenames) # check what variables are there (in the dictionary): 
#filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/ellipse_axis_angle_for_plotting.mat"
filenames = "/Users/North/Drive/Work/UniH_Work/DataAnalysis/Analysis/Matfiles/ellipse_axis_angle_freq_for_plotting.mat"
mat4 = scipy.io.loadmat(filenames) # check what variables are there (in the dictionary): 

drifters2 = OrderedDict() # Preallocate output dictionary
# for keys,drifter in zip(drifters,mat['rbar'].shape[1]):
for drifter in range(mat['rbar'].shape[1]):
    # need to remove nan at the start to get time right
    mask = np.isnan(mat4['lonres'][0,drifter])
    # create time vector 
    date = [dt.datetime.fromordinal(int(ij)) + dt.timedelta(days=ij%1) - dt.timedelta(days = 366) 
            for ij in mat4['t1'][0,drifter][~mask].flatten()] # Convert time axis to date form:
    drifters2['Drifter{}'.format(drifter+1)] = xr.Dataset({'rbar': (['time'], mat['rbar'][0,drifter][~mask]),
                                                           'vbar': (['time'], mat['vbar'][0,drifter][~mask]),
                                                           'vins': (['time'], mat['vins'][0,drifter][~mask]),
                                                           'vazi': (['time'], mat['vazi'][0,drifter][~mask]),
                                                           'vgeo': (['time'], mat4['vgeo'][0,drifter][~mask]),
                                                           'cv': (['time'], (mat['cv'][0,drifter][~mask].real**2+mat['cv'][0,drifter][~mask].imag**2)**0.5),
                                                           'cvhat': (['time'], (mat['cvhat'][0,drifter][~mask].real**2+mat['cvhat'][0,drifter][~mask].imag**2)**0.5),
                                                           'cvres': (['time'], (mat['cvres'][0,drifter][~mask].real**2+mat['cvres'][0,drifter][~mask].imag**2)**0.5),
                                                           'ro': (['time'], mat['ro'][0,drifter][~mask]),
                                                           'lon_matlab': (['time'], mat4['lon'][0,drifter][~mask]),
                                                           'lat_matlab': (['time'], mat4['lat'][0,drifter][~mask]),
                                                           'lon_res': (['time'], mat4['lonres'][0,drifter][~mask]),
                                                           'lat_res': (['time'], mat4['latres'][0,drifter][~mask]),
                                                           'freq_bar': (['time'], mat4['fbar'][0,drifter][~mask]/(2*np.pi)), # cyc/day
                                                           'time_matlab': (['time'], mat4['t1'][0,drifter][~mask]),
                                                           'semi_major_ax': (['time'], mat4['semi_major'][0,drifter][~mask]),
                                                           'semi_minor_ax': (['time'], mat4['semi_minor'][0,drifter][~mask]),
                                                           'theta': (['time'], mat4['the'][0,drifter][~mask]),
                                                           'kappa': (['time'], mat4['kap'][0,drifter][~mask]),
                                                           'lamda': (['time'], mat4['lam'][0,drifter][~mask]),
                                                           'phi': (['time'], mat4['phi'][0,drifter][~mask])},
                                                          coords={'time': date})
    
# merge into one dataset
# not using interp_like as Jlab has some data gaps that should remain
drifters3 = OrderedDict() # Preallocate output dictionary
drifters4 = OrderedDict() # Preallocate output dictionary
for key in drifters:
    drifters3[key] = xr.merge((drifters2[key].interp(time=drifters[key].time),drifters[key])) # match timestamps
    drifters3[key]['DrifterNumber'] = key
    drifters3[key]['velocity'] = (drifters3[key].u**2 + drifters3[key].v**2)**0.5
    drifters3[key]['ke_from_uv'] = 0.5*(drifters3[key].u**2+drifters3[key].v**2) # m**2/s**2
    drifters3[key]['ke_from_vbar'] = 0.5*(drifters3[key].vbar)**2 # m**2/s**2
    drifters3[key]['ke_from_cv'] = 0.5*(drifters3[key].cv)**2 # m**2/s**2
    drifters3[key] = drifters3[key].isel(time=np.arange(1,drifters3[key].u.size,1))  # many first values are very high; remove as outliers
    drifters4[key] = xr.merge([drifters3[key][['vbar','vins','vazi','vgeo','cv','cvhat','cvres','velocity',]].to_array(dim='velocity_type',name='ellipse_velocity'),
                               drifters3[key][['ke_from_uv','ke_from_vbar','ke_from_cv']].to_array(dim='ke_type',name='kinetic_energy'),
                               drifters3[key][['lon_matlab','lat_matlab','lon_res','lat_res','semi_major_ax','semi_minor_ax','theta','vgeo',
                                               'kappa','lamda','phi','time_matlab','freq_bar','rbar','ro',]].to_array(dim='ell_dims',name='ellipse_properties'),
                               drifters3[key][['lon_res','lat_res','u','v','lon','lat','temperature','DrifterNumber']]])
drifters4[key]
```
