#### Plots of all revelevant transects with density (CTD/Scanfish) and currents (LADCP/SADCP) data

```python
#from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import warnings
import xarray as xr
import os
from collections import OrderedDict
from itertools import islice # for slice OrderedDict
from glob import glob
import cartopy.crs as ccrs

os.chdir("/Users/North/Drive/Work/UniH_Work/DataAnalysis/jlab_files") # need to set working directory to get the next step to work
from Benguela.src.met132_ctd_ladcp_functions import load_combine_ladcp_ctd_data #import load_ctd_data
from Benguela.src.met132_scanfish_sadcp_functions import load_combine_sadcp_scanfish_data #import load_ctd_data
import Benguela.src.met132_calc_functions as cf
import Benguela.src.met132_plotting_functions as fg
import Benguela.src.met132_sea_level_functions as sl
import Denmark_Strait.src.spectra_and_wavelet_functions as sw
import Denmark_Strait.src.ssa_core as ssa
```

```python
sadcp_nc_dir = '/Users/North/Drive/Work/UniH_Work/DataAnalysis/jlab_files/Benguela/netcdf_files/'
figure_dir = '/Users/North/Drive/Work/UniH_Work/DataAnalysis/jlab_files/Benguela/figures/paper/'

def read_nc(files,dim_concat,dim):
    # glob expands paths with * to a list of files, like the unix shell
    paths = sorted(glob(files))
    datasets = [xr.open_dataset(p).swap_dims({'xy':dim}) for p in paths]
    combined = xr.concat(datasets,dim_concat)
    return combined

# station and time_secs are in the end somewhat dummy variables but enable transects to be a new dimension
ctd_ladcp_transects = read_nc(sadcp_nc_dir+'ctd_ladcp_transects_LADCP_CTD_Transect*.nc', dim_concat='transect', dim='station')
ctd_ladcp_transects['transect'] = ctd_ladcp_transects.time.min(dim='station',skipna=True) # set transect values to start time
ctd_ladcp_transects = ctd_ladcp_transects.sortby('transect') # sort dataset by transect start time

scan_sadcp_transects = read_nc(sadcp_nc_dir+'scan_sadcp_transects_1km_*.nc',dim_concat='transect', dim='time_secs')
scan_sadcp_transects['transect'] = scan_sadcp_transects.time.min(dim='time_secs',skipna=True) # set transect values to start time
scan_sadcp_transects = scan_sadcp_transects.sortby('transect') # sort dataset by transect start time
    
# ====== load SST map
def read_sst_nc(files,dim_concat):
    paths = sorted(glob(files))
    datasets = [xr.open_dataset(p).assign_coords(Date=xr.open_dataset(p).time_coverage_start[:-14]).sst for p in paths]
    combined = xr.concat(datasets,dim_concat)
    combined['Date'] = combined['Date'].astype("datetime64[ns]")
    return combined 

sst_files = '/Users/North/Drive/Work/UniH_Work/DataAnalysis/Data/MET_132/Remote_Sensing/downloaded_sst_nc_files_on_20.11.2019/A2016*.L3m_DAY_SST.x_sst.nc'
sst_map = read_sst_nc(sst_files,'Date')

# ====== load SSH map
gos_map = xr.open_dataset(sadcp_nc_dir+'sla_gos.nc')
# ====== sadcp for full cruise ship track
sadcp = xr.open_dataset(sadcp_nc_dir+'sadcp_1km.nc')
```

#### CTD/LADCP Filament transects

```python
warnings.filterwarnings('ignore')
plt.rcParams.update({'font.size': 18})

si, nrows, ncols =  0, 4, 4
projection = ccrs.PlateCarree()
x_y_lim = [8.5,15.5,-29.25,-24.]
fig, axs = plt.subplots(nrows=nrows,ncols=ncols,figsize = (ncols*8.,nrows*6))
axs = axs.ravel()

sik = 0
for itransect in np.array((0,1,4,5)):
    si = sik
    # ==== SST map, ship track and transects
    sst_map.sel(Date=ctd_ladcp_transects.isel(transect=itransect).time.dropna(dim='station',how='all')[0].values,method='nearest').plot(ax=axs[si],vmin=14,vmax=20,cmap=plt.cm.coolwarm)
    gos_map.adt.sel(time=ctd_ladcp_transects.isel(transect=itransect).time.dropna(dim='station',how='all')[0].values,method='nearest').plot.contour(ax=axs[si],vmin=0.25,
                                                                                                               vmax=0.4,alpha=0.75)
    axs[si].plot(sadcp.lon.values,sadcp.lat.values,color='0.55',alpha=0.5)
    axs[si].plot(ctd_ladcp_transects.isel(transect=itransect).lon,ctd_ladcp_transects.isel(transect=itransect).lat,
                 color='magenta',linewidth=2)
    ctd_ladcp_transects.isel(transect=itransect).sel(z=0,method='nearest').plot.scatter(x='lon',y='lat',hue='CT',ax=axs[si],vmin=14,vmax=20,
                                                              add_guide=False,cmap=plt.cm.coolwarm)
    axs[si].set_xlim(x_y_lim[0:2])
    axs[si].set_ylim(x_y_lim[2:4])
    axs[si].set_rasterized(True) # !! to reduce file size make this image, rest stays vectors

    si = si+ncols
    # ==== CT transect
    ctd_ladcp_transects.isel(transect=itransect).CT.dropna(dim='station',how='all').plot(x='x_km',ax=axs[si],vmin=15,vmax=18,cmap=plt.cm.coolwarm)
    ctd_ladcp_transects.isel(transect=itransect).sigma_0.dropna(dim='station',how='all').plot.contour(x='x_km',ax=axs[si],
                                                                                                      levels=np.arange(24.2,27.2,0.1),
                                                                                                      colors='w',alpha=0.7)
    axs[si].set_ylim((-100,0))
    axs[si].text(0,0,ctd_ladcp_transects.isel(transect=itransect).dropna(dim='station',how='all').time[0].values.astype('datetime64[D]'),
                   transform=axs[si].transAxes,horizontalalignment='left',verticalalignment='top',fontsize=10)        
    axs[si].text(1,0,ctd_ladcp_transects.isel(transect=itransect).dropna(dim='station',how='all').time[-1].values.astype('datetime64[D]'),
                   transform=axs[si].transAxes,horizontalalignment='right',verticalalignment='top',fontsize=10)        
    axs[si].text(1,1,('Mean dx = '+str(ctd_ladcp_transects.x_km.diff('station').mean().round(1).values)+' km'+
                     ' and mean dz = '+str(ctd_ladcp_transects.z.diff('z').mean().round(1).values)+' m'),
                   transform=axs[si].transAxes,horizontalalignment='right',verticalalignment='top',fontsize=10)        
    ax2 = axs[si].twiny() 
    ax2.set_xlim(axs[si].axis()[0], axs[si].axis()[1])
    ax2.set_xticklabels((ctd_ladcp_transects.isel(transect=itransect).CT.dropna(dim='station',how='all').swap_dims({'station': 'x_km'}).
                         interp(x_km=axs[si].get_xticks()).lat).values.round(1))
    axs[si].set_title('')

    # ==== SA transect
    si = si+ncols
    ctd_ladcp_transects.isel(transect=itransect).SA.dropna(dim='station',how='all').plot(x='x_km',ax=axs[si],vmin=35,vmax=35.6)
    ctd_ladcp_transects.isel(transect=itransect).sigma_0.dropna(dim='station',how='all').plot.contour(x='x_km',ax=axs[si],
                                                                                                      levels=np.arange(24.2,27.2,0.1),
                                                                                                      colors='w',alpha=0.7)
    axs[si].set_ylim((-100,0))
    axs[si].set_title('')

    # ==== U transect
    si = si+ncols
    ctd_ladcp_transects.isel(transect=itransect).u.dropna(dim='station',how='all').plot(x='x_km',ax=axs[si],vmin=-0.4,vmax=0.4,
                                                                                       cmap=plt.cm.PiYG)
    ctd_ladcp_transects.isel(transect=itransect).sigma_0.dropna(dim='station',how='all').plot.contour(x='x_km',ax=axs[si],
                                                                                                      levels=np.arange(24.2,27.2,0.1),
                                                                                                      colors='w',alpha=0.7)
    axs[si].set_ylim((-100,0))
    axs[si].set_title('')
    sik = sik +1

figname = figure_dir+'Supp_Figure_1_CTD_LADCP_Transects_Filament.pdf'
fig.savefig(figname, orientation='landscape', bbox_inches='tight')
```
